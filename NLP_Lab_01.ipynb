# **Remove Special Characters in the given text**

```python
# remove Special Characters in the given text

def remove_special_char(text):
    ans = ""
    for char in text:
        if char.isalnum() or char.isspace():
            ans += char
    return ans

input_text= "this is an example @ text ^ for texing%!"
output_text= remove_special_char(input_text)
print(output_text)
```
