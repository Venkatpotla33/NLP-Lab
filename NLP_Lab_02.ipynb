# **Implement A regular Expression**

```python
import re
plural = 'foods'
match= re.search('s$',plural)
if match is None:
  print('parsing is not sucessful and is not a plural word')
else:
  print('parsing is sucessful and is a plural word')
```

```python
import re

text1 = "hello123world"
text2 = "helloworld"

# Check if the string contains at least one digit
match1 = re.search(r'\d', text1)
match2 = re.search(r'\d', text2)

if match1:
  print(f"'{text1}' contains a digit.")
else:
  print(f"'{text1}' does not contain a digit.")

if match2:
  print(f"'{text2}' contains a digit.")
else:
  print(f"'{text2}' does not contain a digit.")
```

In this example:
- `re.search(r'\d', text)`: This searches for a digit (`\d`) within the given `text`.
- If a digit is found, `re.search` returns a match object; otherwise, it returns `None`.
